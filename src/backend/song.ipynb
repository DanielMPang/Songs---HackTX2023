```python
import pandas as pd
import mpld3
```

```python
import matplotlib.pyplot as plt
```

```python
hit_songs = pd.read_csv("../data/Spotify#1SongsByCountry - May 2020.csv")
```

```python
fig = plt.figure(figsize = (18,8))

titles = hit_songs["Title"]
plt.hist(titles, bins=35, color="lightblue", edgecolor="black", alpha=0.7)
plt.xlabel("Song Title")
plt.ylabel("Frequency")
plt.title("Number of Countries where Songs are #1")
plt.xticks(rotation=90)
plt.show()

html_str = mpld3.fig_to_html(fig)
Html_file= open("index.html","w")
Html_file.write(html_str)
Html_file.close()
```

```python
fig = plt.figure(figsize = (18,8))

artist = hit_songs["Artists"]
plt.hist(artist, bins=35, color="lightblue", edgecolor="black", alpha=0.7)
plt.xlabel("Artist Name")
plt.ylabel("Frequency")
plt.title("Number of Countries where Artists are #1")
plt.xticks(rotation=90)
plt.show()

html_str = mpld3.fig_to_html(fig)
Html_file= open("index.html","w")
Html_file.write(html_str)
Html_file.close()
```

```python
frequent_songs = hit_songs["Title"].value_counts().head(8)
print(frequent_songs)
```

```python
rockstar_countries = songs[songs["Title"] == "ROCKSTAR (feat. Roddy Ricch)"]
print(rockstar_countries)
```

```python
top_rockstar_countries = rockstar_countries[rockstar_countries["Rank"] <= 5]
print(top_rockstar_countries)
```

```python
blinding_countries = songs[songs["Title"] == "Blinding Lights"]
print(blinding_countries)
```

```python
top_blinding_countries = blinding_countries[blinding_countries["Rank"] <= 5]
print(top_blinding_countries)
```

```python
rojo_countries = songs[songs["Title"] == "Rojo"]
print(rojo_countries)
```

```python
top_rojo_countries = rojo_countries[rojo_countries["Rank"] <= 5]
print(top_rojo_countries)
```

```python
safaera_countries = songs[songs["Title"] == "Safaera"]
print(safaera_countries)
```

```python
top_safaera_countries = safaera_countries[safaera_countries["Rank"] <= 5]
print(top_safaera_countries)
```

```python
rain_countries = songs[songs["Title"] == "Rain On Me (with Ariana Grande)"]
print(rain_countries)
```

```python
top_rain_countries = rain_countries[rain_countries["Rank"] <= 5]
print(top_rain_countries)
```

```python
combined_influence_countries = pd.concat([rockstar_countries, blinding_countries, rojo_countries, safaera_countries, rain_countries])
most_influence = combined_influence_countries["Country"].value_counts().head(20)
print(most_influence)
```

```python
combined_top_influence_countries = pd.concat([top_rockstar_countries, top_blinding_countries, top_rojo_countries, top_safaera_countries, top_rain_countries])
most_top_influence = combined_top_influence_countries["Country"].value_counts().head(10)
print(most_top_influence)
```
